## Paso 13: Detectar outliers

**TODO**: Identifica los outliers en el dataset usando el método IQR

```python
# Función para detectar outliers
def detectar_outliers_iqr(data, columna):
    Q1 = data[columna].quantile(0.25)
    Q3 = data[columna].quantile(0.75)
    IQR = Q3 - Q1
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR
    
    outliers = data[(data[columna] < limite_inferior) | (data[columna] > limite_superior)]
    return outliers, limite_inferior, limite_superior

# TODO: Detecta outliers en cada columna
```

## Conclusión final

Este ejercicio demuestra que:

1. **El orden importa**: Escalar antes de imputar puede dar resultados muy diferentes
2. **SimpleImputer vs KNNImputer**: KNN considera las relaciones entre variables pero es más sensible a las escalas
3. **Las escalas afectan**: Variables con diferentes magnitudes pueden dominar el cálculo de distancias en KNN
4. **Los outliers son críticos**: Valores atípicos pueden sesgar significativamente las imputaciones
5. **No hay una solución única**: La elección del enfoque depende del contexto y los datos

### Recomendaciones:
- **Siempre escala antes de usar KNNImputer** cuando las variables tienen diferentes magnitudes
- Explora tus datos antes de imputar
- Detecta y maneja outliers apropiadamente
- Considera usar mediana en lugar de media si hay outliers
- KNNImputer con escalado suele dar mejores resultados cuando hay correlaciones entre variables
